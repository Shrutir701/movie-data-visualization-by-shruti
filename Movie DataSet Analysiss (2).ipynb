*DATA VISUALIZATION*

```python
#BOX PLOT OF RATINGS BY GENRE
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data to create the DataFrame
data = {
    'genere': ['Action', 'Comedy', 'Drama', 'Action', 'Comedy', 'Drama'],
    'rating': [8.5, 7.0, 9.0, 7.5, 6.5, 8.0]
}

# Define the DataFrame
df = pd.DataFrame(data)

# Now you can create the boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='genere', y='rating', data=df)
plt.title('Box Plot of Ratings by Genre')
plt.xticks(rotation=45)
plt.show()
```

**FINDINGS FROM THE ABOVE BOX PLOT OF RATINGS BY GENRE**

1.From the plotting visual we can clearly see that the DRAMA GENRE/MOVIES have the highest median rating indicating that they are generally the most well received genre compared to others.

2.ACTION GENRE/MOVIES rank second in terms of median rating,showing that they are also popular, though their ratings exhibit slightly more variability compared to drama.

3.COMEDY GENRE/MOVIES rank third in ratings,with a wider range of values,suggesting that while some comedies are highly rated, others perform less consistently.

```python
#CORRELATION HEATMAP
import pandas as pd
df=pd.DataFrame({
    'column1': [1,2,3],
    'column2':['A','B','C']
})
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 8))
correlation_matrix = numeric_df.corr()  # Calculate correlation on numeric data
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#BAR PLOT FOR TOP 10 DIRECTORS BY NUMBER OF MOVIES
import warnings
warnings.filterwarnings("ignore")
top_directors = data['Director'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, hue=None, legend=False, palette='magma')
plt.title('Top 10 Directors by Number of Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Directors')
plt.show()
```

***FINDINGS FROM THE ABOVE BAR PLOT***

1.From the given bar plot we can see the names of TOP 10 DIRECTORS in the movie dataset.

2.Out of all 10 Directos,Ridley Scott holds the top spot with the highest number of movies directed,indicating a prolific career and strong industry influence.

3.David Yates, M.Night Shyamalan, Paul W.S. Anderson follows closely, showcasing consistent output and popularity in the industry.

4.Michael Bay,Zack Snyder, Denis Villeneuve, Woody Allen, Peter Berg and Danny Boyle suggests that, they are also highly active in the industry, but with slightly fewer films than the top 5.



```python
#BAR PLOT FOR TOP 10 ACTORS BY NUMBER OF MOVIES
top_stars = data['Actors'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_stars.values, y=top_stars.index, palette='plasma')
plt.title('Top Actors by Number of Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Actors')
plt.show()
```
